```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
#selecting specific country and measure

select_location = 'CAN'
select_measure = 'THND_TONNE'
```

```python
#extracting the sub-data frame for given location and given measure

df_select = df[(df.LOCATION == select_location) & (df.MEASURE ==select_measure)]
```

```python
df_select
```

```python
df_select.LOCATION.unique()
```

```python
#plotting figures for all four crops (subjects) for given measure and location

plt.figure(figsize=(12,6))
sns.lineplot(data=df_select, x='TIME', y='Value', hue = 'SUBJECT', hue_order = all_subjects)
plt.title (select_location + '-' + select_measure)
plt.grid()
plt.show()
```

# Lineplot for each location in the data with all crops measured in thousand tonnes

```python
#looping over all locations

for select_location in all_locations:
    df_select = df[(df.LOCATION == select_location) & (df.MEASURE == select_measure)]
    plt.figure(figsize=(12,5))
    sns.lineplot(data = df_select, x = 'TIME', y = 'Value', hue = 'SUBJECT', hue_order = all_subjects)
    plt.title(select_location + '-' + select_measure)
    plt.grid()
    plt.show()
```
